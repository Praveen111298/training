```python
from PIL import Image
from sklearn.cluster import KMeans
```

```python
image = Image.open('tiger.png')
```

```python
from matplotlib import pyplot as plot

%matplotlib inline

plot.imshow(image)
```

```python
import numpy as np

image_np = np.asarray(image)
```

```python
image_np.shape
```

```python
image_np = image_np.reshape((720*1280,3))
```

```python
image_np.shape
```

```python
kmeans = KMeans(n_clusters=16, random_state=0)
kmeans.fit(image_np)
```

```python
labels = kmeans.predict(image_np)
```

```python
labels = labels.reshape((720,1280))
```

```python
plot.imshow(labels)
```

```python
centers = [centers.astype(int) for centers in kmeans.cluster_centers_]
```
