```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
df=pd.read_csv('Z:\FA1\DataSets\Trees.csv')
a=df.drop(columns='Unnamed: 0',axis=1)
a.plot(kind='box')
```

```python
import matplotlib.pyplot as plt
fig, (fig1,fig2,fig3) = plt.subplots(1, 3)
fig.set_figwidth(10)
fig.set_figheight(7)
fig1.boxplot(df["Girth"],vert=True,notch=True,bootstrap=10)
fig2.boxplot(df["Height"],vert=True,notch=True)
fig3.boxplot(df["Volume"],vert=True,notch=True)
fig1.set_title('Girth of trees')
fig1.set_ylabel('Trees diameter in inches')
fig2.set_title('Height of trees')
fig2.set_ylabel("Trees height in feet ")
fig3.set_title('Volume of trees')
fig3.set_ylabel('Trees height in cubic feet')
fig.show()
```
