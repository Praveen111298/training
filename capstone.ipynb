```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_table("Z:/FA1/DataSets/AdultDataset/Adult Dataset/adult.data",sep=",")
data.columns=["age","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","salary"]
data
```

```python
data.columns
```

```python
data.describe()
```

```python
data1=data.replace(' ?' , np.nan)
data1
```

```python
data1.dropna()
```

```python
data1.describe()
```

```python
aq1=data1["age"].quantile(0.25)
aq3=data1["age"].quantile(0.75)
aiqr=(aq3-aq1)
print(aq1,aq3,aiqr)
```

```python
data1=data1[((data1["age"]> (aq1 - 1.5 * aiqr)) & (data1["age"]< (aq3 + 1.5 * aiqr)))]
```

```python
data1.shape
```

```python
fq1=data1["fnlwgt"].quantile(0.25)
fq3=data1["fnlwgt"].quantile(0.75)
fiqr=(fq3-fq1)
print(fq1,fq3,fiqr)
```

```python
data1=data1[((data1["fnlwgt"]>(fq1 - 1.5 * fiqr)) & (data1["fnlwgt"]< (fq3 + 1.5 * fiqr)))]
```

```python
data1.shape
```

```python
eq1=data1["education-num"].quantile(0.25)
eq3=data1["education-num"].quantile(0.75)
eiqr=(eq3-eq1)
print(eq1,eq3,eiqr)
```

```python
data1=data1[((data1["education-num"]> (eq1 - 1.5 * eiqr)) & (data1["education-num"]< (eq3 + 1.5 * eiqr)))]
data1.shape
```

```python
gq1=data1["capital-gain"].quantile(0.25)
gq3=data1["capital-gain"].quantile(0.75)
giqr=(gq3-gq1)
print(gq1,gq3,giqr)
```

```python
data1=data1[((data1["capital-gain"]>=(gq1 - 1.5 * giqr)) & (data1["capital-gain"]<= (gq3 + 1.5 * giqr)))]
```

```python
data1.shape
```

```python
lq1=data1["capital-loss"].quantile(0.25)
lq3=data1["capital-loss"].quantile(0.75)
liqr=(lq3-lq1)
print(lq1,lq3,liqr)
```

```python
data1=data1[((data1["capital-loss"]>= (lq1 - 1.5 * liqr)) & (data1["capital-loss"]<= (lq3 + 1.5 * liqr)))]
data1.shape
```

```python
hq1=data1["hours-per-week"].quantile(0.25)
hq3=data1["hours-per-week"].quantile(0.75)
hiqr=(hq3-hq1)
print(hq1,hq3,hiqr)
data1.shape
```

```python
data1=data1[((data1["hours-per-week"]> (hq1 - 1.5 * hiqr)) & (data1["hours-per-week"]< (hq3 + 1.5 * hiqr)))]
data1.shape
```

```python
a=data1.groupby(["salary","hours-per-week"]).count()["capital-gain"].unstack()
a
```

```python
b=a.T
b
```
