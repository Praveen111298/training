```python
# 3 c min max normalization it is done when range between the values are high 
from  sklearn import preprocessing
scaler =preprocessing.MinMaxScaler()
data1=pd.DataFrame(scaler.fit_transform(pima_df))
data1.columns=["Pregnancies","Glucose","Blood Pressure","Skin Thickness","Insulin","BMI" ,"Diabetes Pedigree Function","Age","Class"]
data1.head()
```
