```python
from sklearn.preprocessing import MinMaxScaler
k=MinMaxScaler()
```

```python
x=k.fit_transform(df.drop(labels=['name'],axis=1)) 
x=pd.DataFrame(x,columns=['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP',
       'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR', 'status', 'RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE'])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
f=x[['MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP',
       'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR', 'RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE']]
t=x['status']
```

```python
med_train,med_test,label_train,label_test=train_test_split(f,t,test_size=0.2,random_state=0)
```

```python
from sklearn.decomposition import PCA
```

```python
lda=PCA(0.95)
```

```python
med_train=lda.fit_transform(med_train,label_train)
med_test=lda.transform(med_test)
```

```python
mod=LogisticRegression()
mod.fit(med_train,label_train)


# predicting the test set results

y_pred=mod.predict(med_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
cm=confusion_matrix(label_test,y_pred)
display(cm)
```

```python
accuracy_score(label_test,y_pred)
```

```python
print("train accuracy")
display(mod.score(med_train,label_train))


print("test accuracy")
display(mod.score(med_test,label_test))
lr=mod.score(med_test,label_test)
```

```python
train_prediction=mod.predict(med_train)
test_prediction=mod.predict(med_test)
train_conf_matrix=confusion_matrix(label_train,train_prediction)
test_conf_matrix=confusion_matrix(label_test,test_prediction)
```

```python
pd.DataFrame(train_conf_matrix,columns=mod.classes_,index=mod.classes_)
```

```python
pd.DataFrame(test_conf_matrix,columns=mod.classes_,index=mod.classes_)
```

```python
print(classification_report(label_train,train_prediction))
```

```python
print(classification_report(label_test,test_prediction))
```

```python
# decision tree algorithm 

model=DecisionTreeClassifier()

model.fit(med_train,label_train)
print("train accuracy")
display(model.score(med_train,label_train))

print("test accuracy")
display(model.score(med_test,label_test))
dt=model.score(med_test,label_test)
```

```python
y_pred=model.predict(med_test)
y_pred
```

```python
cm=confusion_matrix(label_test,y_pred)
display(cm)
```

```python
# using min_sample_split and min_impurity_decrease

model1=DecisionTreeClassifier(min_samples_split=10,min_impurity_decrease=0.005)

model1.fit(med_train,label_train)
print("train accuracy")
display(model1.score(med_train,label_train))

print("test accuracy")
display(model1.score(med_test,label_test))

dt1=model1.score(med_test,label_test)
```

```python
model2=DecisionTreeClassifier(min_samples_split=20,min_impurity_decrease=0.1)


model2.fit(med_train,label_train)
print("train accuracy")
display(model2.score(med_train,label_train))

print("test accuracy")
display(model2.score(med_test,label_test))
dt2=model2.score(med_test,label_test)
```

```python
# k nearest algorithm    


kmodel=KNeighborsClassifier(n_neighbors=3,metric="euclidean")


kmodel.fit(med_train,label_train)
print("train accuracy")
display(kmodel.score(med_train,label_train))

print("test accuracy")
display(kmodel.score(med_test,label_test))
knn1=kmodel.score(med_test,label_test)
```

```python
# changing the n_neighbors value to 5 and validating it


k1model=KNeighborsClassifier(n_neighbors=5,metric="euclidean")
k1model.fit(med_train,label_train)
print("train accuracy")
display(k1model.score(med_train,label_train))

print("test accuracy")
display(k1model.score(med_test,label_test))
knn2=k1model.score(med_test,label_test)
```

```python
smodel=SVC(kernel="rbf")
smodel.fit(med_train,label_train)
print("train accuracy")
display(smodel.score(med_train,label_train))

print("test accuracy")
display(smodel.score(med_test,label_test))

svm=smodel.score(med_test,label_test)
```

```python
## RandomForestClassifier

from sklearn.ensemble import RandomForestClassifier
rmodel=RandomForestClassifier(n_estimators=10,min_samples_split=20,min_impurity_decrease=0.1)
rmodel.fit(med_train,label_train)
print("train accuracy")
display(rmodel.score(med_train,label_train))

print("test accuracy")
display(rmodel.score(med_test,label_test))
rm=rmodel.score(med_test,label_test)
```

```python
from sklearn.ensemble import AdaBoostClassifier
amodel=AdaBoostClassifier(n_estimators=10)
amodel.fit(med_train,label_train)
print("train accuracy")
display(amodel.score(med_train,label_train))

print("test accuracy")
display(amodel.score(med_test,label_test))

ab=amodel.score(med_test,label_test)
```
